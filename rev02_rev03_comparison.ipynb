```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from glob import glob
import os, sys
from os import path
import re
```

```python
rev02_stats = r'C:\Users\yuhan\work\chime\rev02_stats'
rev03_stats = r'C:\Users\yuhan\work\chime\rev03_stats'
fl = 'CSD_stats_out_fr{freq}.csv'
rev02_files = glob(path.join(rev02_stats, fl.format(freq = '*')))
rev03_files = glob(path.join(rev03_stats, fl.format(freq = '*')))
```

```python
def get_freq(file_name):
    """Return frequency according to file name"""
    nums = re.findall(r'\d+', file_name)  # Find regular expression of any non-zero number of digits
    if len(nums) == 2: # "nums" are in the format ['03', freq]
        return int(nums[1])
    else:
        return None
```

```python
rev02 = {get_freq(f) : f for f in rev02_files}
rev03 = {get_freq(f) : f for f in rev03_files}

dat_rev02 = pd.read_csv(rev02[107], header = 0, sep = ',')  # 758 MHz
dat_rev03 = pd.read_csv(rev03[107], header = 0, sep = ',')  # 758 MHz
```

```python
for dat in (dat_rev02, dat_rev03):
    mean = dat['mean']
    stdev = dat['stdev']
    delta = np.square(stdev) + np.square(mean)  # E[X^2], square of the differenced map
    dat.insert(loc=3, column='delta', value=delta)  # Insert an extra column to the DataFrame
```

```python
dat_rev02['type'] = ['rev02' for i in range(dat_rev02.shape[0])]
dat_rev03['type'] = ['rev03' for i in range(dat_rev03.shape[0])]
```

```python
joined = pd.concat([dat_rev02, dat_rev03])
```
