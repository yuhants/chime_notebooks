```python
def get_freq(file_name):
    """Return frequency according to file name"""
    nums = re.findall(r'\d+', file_name)  # Find regular expression of any non-zero number of digits
    if (nums[0] == '03' and len(nums)==2): # "nums" are in the format ['03', freq]
        return int(nums[1])
    else:
        return None
```

```python
class Data:
    def __init__(self, dat):
        self.data = dat
```

```python
stats = {get_freq(f) : f for f in stats_files}  # Dictionary, key = frequency
dat_107 = pd.read_csv(stats[107], header = 0, sep = ',')  # 758 MHz
dat_244 = pd.read_csv(stats[244], header = 0, sep = ',')  # 665 MHz
dat_619 = pd.read_csv(stats[619], header = 0, sep = ',')  # 558 MHz
dat_936 = pd.read_csv(stats[936], header = 0, sep = ',')  # 434 MHz
```

```python
for idx in [107, 244, 619, 936]:
    dat = globals()['dat_'+str(idx)]
    mean = dat['mean']
    stdev = dat['stdev']
    delta = np.square(stdev) + np.square(mean)  # E[X^2], square of the differenced map
    dat.insert(loc=3, column='delta', value=delta)  # Insert an extra column to the DataFrame
    globals()['Data_'+str(idx)] = Data(dat)
```

## Identify outliers

```python
sns.reset_orig()

def histogram(data, bins='fd', range=None):
    """Compute the histogram of a dataset"""
    data = data[~np.isnan(data)]
    hist, bin_edges = np.histogram(data, bins, range, density=False)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2.
    return hist, bin_centers
```

```python
def plot_hist(ax, data, bins=100, range=None, title=None, ylabel=True):
    ax.hist(data, color = 'k', histtype = 'step', 
             bins=bins, range=range, density = False)
    ax.set_ylabel('Count')
    
    if title is not None:
        ax.set_title(title)
        
    if ylabel == False:
        ax.set_ylabel('')
```

In previous work people fit gaussian to the binned histogram. This approach is intuitive but nonideal. Because the fitting depends on the choice of binning scheme, we cannot consistently estimate the "location" and "scale" of the distribution.

```python
from scipy.optimize import curve_fit

def gauss(x, sigma, mu, N):
    return N * np.exp(- 0.5 * ((x-mu)/sigma)**2)

def gaussian_fit(data):
    n, mid_bin = histogram(data)
    (sigma, mu, N), pcov = curve_fit(gauss, mid_bin, n)
    return sigma, mu, N
```

```python
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111)
plot_hist(ax, mean, 'fd', title='Mean')

n, mid_bins = histogram(mean)
x = np.linspace(mid_bins[0], mid_bins[-1], 1000)
sigma, mu, N = gaussian_fit(mean)
ax.plot(x, gauss(x, sigma, mu, N), color='green', label='Gaussian Fit')
print('mu:', mu)
print('sigma:', sigma)
print(N)
```

A better way is to simply calculate mean and standard deviation - they are really what we get if least-square "fitting" the unbinned data. Or put it in another way, the maximum likelihood solution for a guassian likelihood.

```python
_ = mean[~np.isnan(mean)]
print('Mean:', np.mean(_))
print('STD:', np.std(_))
```

However, mean and std are not robust in the presence of outliers (can be seen from their deviation from values given by histogram fitting). An alternative way is using median and interquartile range (IQR).

A conventional "definition" for outliers:

data outside of the range (lower, upper) = (q25 - 1.5 $\times$ IQR, q75+1.5 $\times$ IQR)

This correspond to ~2.7 sigma for a normal distribution

```python
q75, q25 = np.percentile(_, [75 ,25])
iqr = q75 - q25
print('Median', np.median(_))
print('Normalized Interquartile Range:', 0.7413 * iqr)
```

```python
def get_outliers(dat, stat, method='iqr'):
    """Return outliers in terms of CSDs"""
    data = dat.dropna()
    
    if method == 'iqr':
        _ = data[stat].to_numpy()
        
        q75, q25 = np.percentile(_, [75 ,25])
        iqr = q75 - q25
        lower, upper = q25 - 1.5*iqr, q75 + 1.5*iqr
        outliers_l = dat['CSD'][dat[stat] < lower]
        outliers_u = dat['CSD'][dat[stat] > upper]
        
        return outliers_l.to_numpy(), outliers_u.to_numpy()

    elif method == 'count':
        size = int(len(data.index) * 0.05)  # Pick the largest 5%
        sorted_data = data.sort_values(by=stat, ascending=False, inplace=False)
        outliers = sorted_data['CSD'][:size]
        
        return outliers.to_numpy()
```

```python
for idx in [107, 244, 619, 936]:
    Data = globals()['Data_'+str(idx)]
    for stat in ['mean', 'median', 'skew']:
        setattr(Data, f'out_{stat}', np.sort(np.concatenate(get_outliers(Data.data, stat, 'iqr'))))

    for stat in ['delta', 'stdev', 'kurtosis', 'zsum']:
        setattr(Data, f'out_{stat}', np.sort(get_outliers(Data.data, stat, 'count')))
```

```python
def plot_outliers(Data, title='Outliers'):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111)
    
    count = 0
    for stat in ['mean', 'median', 'skew', 'kurtosis', 'delta', 'stdev', 'zsum']:
        o = getattr(Data, f'out_{stat}')
        ax.barh(o, np.ones_like(o), left=count, label=stat)
        count += 1
    ax.set_title(title, fontsize=24, pad=10)
    ax.set_xticklabels('')

    ax.set_xticks([0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5], minor=True)
    ax.set_xticklabels(['mean', 'median', 'skew', 'kurtosis', 'delta', 'stdev', 'zsum'], minor=True)

    ax.set_ylabel('CSD', fontsize=20, labelpad=10)
    ax.tick_params(axis='x', which='minor', labelsize=18)
    ax.tick_params(axis='y', labelsize=18)
    ax.xaxis.grid(linestyle='--')
```

```python
plot_outliers(Data_107, title='Outliers: 758 MHz')
plot_outliers(Data_244, title='Outliers: 665 MHz')
plot_outliers(Data_619, title='Outliers: 558 MHz')
plot_outliers(Data_936, title='Outliers: 434 MHz')
```
